# Holoborodko filters
This notebook is intended to provide additional information and implementations of the smooth noise-robust differentiation method from Holoborodko that was used in the baseline correction method presented in the other notebook in this repository (https://github.com/cstckr/whittaker_baseline_correction/blob/main/whittaker_baseline_correction.ipynb).

Reference:
- Holoborodko, Smooth noise robust differentiators. Online: http://www.holoborodko.com/pavel/numerical-methods/numerical-derivative/smooth-low-noise-differentiators/

We will use the same data as test data as for the baseline correction method:

```python
import numpy as np
import matplotlib.pyplot as plt

def lorentz(a, c, w, x):
    y = a * (1 / (1 + ((x - c) / (0.5 * w))**2))
    return y


x = np.linspace(-0.5, 1, 512)
np.random.seed(42)
sp_noise = np.random.normal(0, 1, size=len(x))
sp_baseline = (-x**3 + x**2) * 40 + 5
sp_signals = lorentz(25, 0.24, 0.06, x) + lorentz(30, 0.35, 0.08, x)
y = sp_noise + sp_baseline + sp_signals
```

The smooth noise-robust differentiation proposed by Holoborodko gives similiar results for small filter lengths compared to Savitzky-Golay filtering (which has an added y-offset in the graphic below for clarity). However it is apparent that for the Savitzky-Golay filter significant over-smoothing is quickly obtained when the filter length is increased:

```python
from scipy.special import binom

def robust_derivative(y, N):
    try:
        if (N % 2 == 0) | (N < 5):
            raise ValueError("Filter length N must be odd and at least 5.")
    except ValueError as error:
        print(error)
       
    m = (N - 3) // 2
    M = (N - 1) // 2
    d = 0
    window = np.lib.stride_tricks.\
             sliding_window_view(y, window_shape=len(y) - M * 2)
    for k in range(1, M + 1):
        c_k = 1 / 2**(2 * m + 1) * \
              (binom(2 * m, m - k + 1) - \
               binom(2 * m, m - k - 1))
        d += c_k * (window[M + k] - window[M - k])
    d = np.pad(d, pad_width=(M, M), mode="constant", constant_values=np.nan)
    return d


from scipy.signal import savgol_filter

d1 = robust_derivative(y, N=15)
d2 = savgol_filter(y, 15, 3, deriv=1) + 3

fig, ax = plt.subplots()
ax.plot(x, y, color="k", label="Spectrum")
ax.plot(x, d1, color="r", label="Robust derivative")
ax.plot(x, d2, color="b", label="Savitzky-Golay derivative")
ax.invert_xaxis()
plt.xlabel("x"), plt.ylabel("y")
ax.legend(loc="lower center", bbox_to_anchor=(0, 1.05, 1, 0.2), ncol=4)
plt.show()
```

```python
d1 = robust_derivative(y, N=75)
d2 = savgol_filter(y, 75, 3, deriv=1) + 3

fig, ax = plt.subplots()
ax.plot(x, y, color="k", label="Spectrum")
ax.plot(x, d1, color="r", label="Robust derivative")
ax.plot(x, d2, color="b", label="Savitzky-Golay derivative")
ax.invert_xaxis()
plt.xlabel("x"), plt.ylabel("y")
ax.legend(loc="lower center", bbox_to_anchor=(0, 1.05, 1, 0.2), ncol=4)
plt.show()
```

# Noise-robust smoothing without differentiation
The method can be extended to smooth signals without differentiation, see the code blocks below:

Reference: 
Andrey Paramonov, http://www.holoborodko.com/pavel/numerical-methods/noise-robust-smoothing-filter/

```python
def robust_smoother(y, N):
    try:
        if (N % 2 == 0) | (N < 5):
            raise ValueError("Filter length N must be odd and at least 5.")
    except ValueError as error:
        print(error)
       
    m = (N - 1) // 2
    f = 0
    window = np.lib.stride_tricks.\
             sliding_window_view(y, window_shape=len(y) - m * 2)
    for k in range(-m, m + 1):
        c_k = ((3 * m - 1 - 2 * k**2) / (2 * m - 1)) * \
              (1 / 2**(2 * m)) * \
              binom(2 * m, m + k)
        f += c_k * window[k + m]
    f = np.pad(f, pad_width=(m, m), mode="constant", constant_values=np.nan)
    return f


f1 = robust_smoother(y, 31)

fig, ax = plt.subplots()
ax.plot(x, y, color="k", label="Spectrum")
ax.plot(x, f1, color="r", label="Smoothed")
ax.invert_xaxis()
plt.xlabel("x"), plt.ylabel("y")
ax.legend(loc="lower center", bbox_to_anchor=(0, 1.05, 1, 0.2), ncol=4)
plt.show()
```

# Smooth noise-robust second derivative
The method can also be extended to obtain smooth second derivatives, see the code blocks below. However, in the implementation below computational overhead becomes significant for larger filter lengths.

```python
def robust_2nd_derivative(y, N):
    try:
        if (N % 2 == 0) | (N < 5):
            raise ValueError("Filter length N must be odd and at least 5.")
    except ValueError as error:
        print(error)
    M = (N - 1) // 2

    def s(k):
        if k > M:
            return 0
        elif k == M:
            return 1
        else:
            return ((2 * N - 10) * s(k + 1) - \
                    (N + 2 * k + 3) * s(k + 2)) / \
                    (N - 2 * k - 1)

    window = np.lib.stride_tricks.\
             sliding_window_view(y, window_shape=len(y) - M * 2)
    sum_term = 0
    for k in range(1, M + 1):
        sum_term += s(k) * (window[M + k] + window[M - k])
    d = (1 / 2**(N - 3)) * (s(0) * window[M + 1] + sum_term)
    d = np.pad(d, pad_width=(M, M), mode="constant", constant_values=np.nan)
    return d


spectrum_wo_noise = sp_baseline + sp_signals 
diff2_wo_noise = np.diff(spectrum_wo_noise, n=2)
diff2_wo_noise = np.pad(diff2_wo_noise, pad_width=(1, 1), mode="constant", constant_values=np.nan)

d3 = robust_2nd_derivative(y, 31)

fig, ax = plt.subplots()
ax.plot(x, diff2_wo_noise, color="k", label="2nd derivative without noise")
ax.plot(x, d3, color="r", label="Smoothed")
ax.invert_xaxis()
plt.xlabel("x"), plt.ylabel("y")
ax.legend(loc="lower center", bbox_to_anchor=(0, 1.05, 1, 0.2), ncol=4)
plt.show()
```
